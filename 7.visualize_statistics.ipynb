```python
import pandas as pd
import matplotlib.pyplot as plt

def main():
    # 1) LOAD THE PROCESSED FEATURE DATA
    input_csv = "data/mar_11_features_3class.csv"
    df = pd.read_csv(input_csv)
    print(f"Loaded {len(df)} rows from {input_csv}")
    print(df.head())
    
    # 2) EXAMPLE PLOT: LINE CHART OF ACC_MEAN OVER start_time
    plt.figure()  # separate figure
    plt.plot(df["start_time"], df["acc_mean"])
    plt.xlabel("Start Time (s)")
    plt.ylabel("ACC Mean")
    plt.title("ACC Mean vs. Start Time")
    plt.show()

    # 3) EXAMPLE PLOT: LINE CHART OF GYRO_MEAN OVER start_time
    plt.figure()  # separate figure
    plt.plot(df["start_time"], df["gyro_mean"])
    plt.xlabel("Start Time (s)")
    plt.ylabel("GYRO Mean")
    plt.title("GYRO Mean vs. Start Time")
    plt.show()

    # 4) EXAMPLE PLOT: HISTOGRAM OF ACC_MEAN
    plt.figure()  # separate figure
    plt.hist(df["acc_mean"].dropna(), bins=30)
    plt.xlabel("ACC Mean")
    plt.ylabel("Frequency")
    plt.title("Distribution of ACC Mean")
    plt.show()

    # 5) EXAMPLE PLOT: BAR CHART OF LABEL COUNTS
    #    (For example, how many FALL vs. NO_FALL)
    plt.figure()  # separate figure
    label_counts = df["label"].value_counts()
    plt.bar(label_counts.index.astype(str), label_counts.values)
    plt.xlabel("Label")
    plt.ylabel("Count")
    plt.title("Label Distribution")
    plt.show()

    print("Visualization complete.")

if __name__ == "__main__":
    main()
```
